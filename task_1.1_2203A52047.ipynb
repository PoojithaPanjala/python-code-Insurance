```python
limit= df[(df['age'] >= 20) & (df['age'] <= 40)]
age=limit['age']
smoker=limit['smoker']
plt.scatter(age, smoker)
plt.title("scatter plot")
plt.xlabel("age")
plt.ylabel("smoker")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

non_smokers = df[df['smoker'] == 'no']
non_smoker_expenses = non_smokers['expenses']
plt.figure(figsize=(8, 6))
plt.boxplot([ non_smoker_expenses], labels=['Non-Smokers'])

plt.ylabel('Expenses')
plt.show()
```

```python
import matplotlib.pyplot as plt
li1 = df[(df['sex'] == 'male') & (df['smoker'] == 'yes')]
li2 = df[(df['sex'] == 'female') & (df['smoker'] == 'yes')]
plt.hist(li1['age'], alpha=0.5, label='Male Smokers')
plt.hist(li2['age'], alpha=0.5, label='Female Smokers')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
